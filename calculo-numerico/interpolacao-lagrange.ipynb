```python
def interpola(x, y):
    
    X = np.vander(x)
    p = al.solve(X, y)
    
    return p.T
```

```python
def lagrange_interp(xi, yi, x):
    
    n = len(xi)
    
    xi, yi, x = np.array(xi), np.array(yi), np.array(x)
    
    if n == 1 :
        
        xi = xi.T
        yi = yi.T
        x = x.T
        
    L = np.ones((n, len(x)))
    
    for i in np.arange(n):
        for j in np.arange(n):
            
            if i != j:
                
                L[i] *= (x - xi[j])/(xi[i] - xi[j])
    
    return yi.dot(L)
```

```python
x = [-2, 0, 3, 5]
y = [3, -2, 4, 2]
```

```python
p = interpola(x, y)
```

```python
xl = np.arange(x[0]-1, x[3]+1, 0.1)
yl = np.polyval(p, xl)

yll = lagrange_interp(x, y, xl)
```

```python
fig, axis = plt.subplots(1, 2, figsize = (12, 6))


plt.sca(axis[0])

plt.title('Interpolação Por Sistemas Lineares', fontsize = 16)

plt.plot(x, y, 'o', label = 'Dados')
plt.plot(xl, yl, label = 'Interpolação')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.sca(axis[1])

plt.title('Interpolação pelo método de Lagrange', fontsize = 16)

plt.plot(x, y, 'o', label = 'Dados')
plt.plot(xl, yll, label = 'Interpolação')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()
```
