```python
def mmq(x, y, k):
    
    n = len(x)
        
    X = np.vander(x)
    X = X[:, n - k - 1: n]
    
    a = al.solve(X.T.dot(X), X.T.dot(y.T))

    return a.T
```

```python
def mmq_qr(x, y, k):
    
    n = len(x)
        
    X = np.vander(x)
    X = X[:, n - k - 1: n]
    
    Q, R = al.qr(X)
    
    a = al.solve(R, Q.T.dot(y.T))

    return a.T
```

```python
p3 = lambda x: 1 + 2*x**(2) - 3*x**(3)
```

```python
x = np.arange(-1, 1, 0.1)
y = p3(x)

r = np.random.normal(0, 1, len(x))
```

```python
p31 = mmq(x, y + r, 3)
p32 = mmq_qr(x, y + r, 3)
```

```python
xl = np.arange(-1, 1, 0.1)

yl1 = np.polyval(p31, xl)
yl2 = np.polyval(p32, xl)
```

```python
fig, axis = plt.subplots(1, 2, figsize = (12, 6))


plt.sca(axis[0])

plt.title('Mínimos Quadrados', fontsize = 16)

plt.plot(x, y, label = 'P(x)')
plt.plot(x, y+r, 'o', label = 'Dados')

plt.plot(xl, yl1, label = 'mmq')

plt.xlabel('x')
plt.ylabel('P(x)')

plt.legend()

plt.sca(axis[1])

plt.title('Mínimos Quadrados por decomposição QR', fontsize = 16)

plt.plot(x, y, label = 'P(x)')
plt.plot(x, y+r, 'o', label = 'Dados')

plt.plot(xl, yl2, label = 'mmq_qr')

plt.xlabel('x')
plt.ylabel('P(x)')

plt.legend()
```
