```python
def euler(f, x0, y0, N, h):
    
    X = np.zeros(N)
    Y = np.copy(X)
    
    X[0] = x0
    Y[0] = y0

    for i in np.arange(N-1):
        
        Y[i+1] = Y[i] + h*f(X[i], Y[i])
        X[i+1] = X[i] + h

    return X, Y
```

```python
def rk4(f, x0, y0, N, h):
    
    X = np.zeros(N)
    Y = np.copy(X)
    
    X[0] = x0
    Y[0] = y0
    
    for i in np.arange(N-1):
        
        k1 = f(X[i], Y[i])
        k2 = f(X[i] + 0.5*h, Y[i] + 0.5*h*k1)
        k3 = f(X[i] + 0.5*h, Y[i] + 0.5*h*k2)
        k4 = f(X[i] + h, Y[i] + h*k3)
        
        Y[i+1] = Y[i] + h*(k1 + 2*(k2 + k3) + k4)/6
        X[i+1] = X[i] + h
        
    return X, Y
```

```python
f = lambda x, y: -y + x + 2
y0 = 2
x0 = 0
N = 10
h = 0.2
```

```python
y = lambda x: np.exp(-x) + x + 1
```

```python
xe, ye = euler(f, x0, y0, N, h)
xr, yr = rk4(f, x0, y0, N, h)
```

```python
plt.figure(figsize=(8,8))

plt.title('Comparação entre métodos de solução de EDOS', fontsize = 16)

plt.plot(xe, y(xe), label = 'y(x)')
plt.plot(xe, ye, '-o', label = 'Euler')
plt.plot(xr, yr, 'o', label = 'Runge-Kutta 4')

plt.legend()
```
