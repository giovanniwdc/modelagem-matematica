
# **Compute Discret Fourier Transform (DFT)**

```python
n = 1024
w = np.exp(-1j* 2 * np.pi / n)
```

```python
J, K = np.meshgrid(np.arange(n), np.arange(n))

DFT = np.power(w, J*K)
DFT = np.real(DFT)
```

```python
fig = plt.figure(figsize = (10,10))
plt.imshow(DFT)
```
